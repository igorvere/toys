```python
import numpy as np
niter=int(1e4)
ndays = 365
dt = ndays/niter
n0=1.0# population
v0=50e3/50e5 #infected population
alpha=0.2#transmission rate per day
gamma=1.0/10#sickness length

v = np.zeros(niter)
v[0] = v0
n = np.zeros(niter)
n[0] = n0

for i in range(0, niter-1):
    n[i+1] = n[i] - alpha * v[i] * n[i] * dt
    v[i+1] = v[i] + alpha * v[i] * n[i] * dt - gamma * v[i] * dt

import matplotlib.pyplot as plt
tt = np.array(range(niter)) * dt
plt.plot(tt, n)
plt.plot(tt, v)
plt.plot(tt, n0-v-n)
plt.legend(['non infected', 'infected', 'immune'])
```
